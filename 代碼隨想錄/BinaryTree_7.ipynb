## 代码
需要比较根节点的**左子树**与**右子树**是不是相互翻转的。（并非是左右节点）
<img src = "https://img-blog.csdnimg.cn/20210203144624414.png">

这道题只能使用【后序遍历】，要通过递归函数的返回值判断两个子树的内侧节点和外侧节点是否相等。

- 单层递归的逻辑=处理左右节点都不为空且数值相同的情况。
  - 比较二叉树外侧是否对称，传入的是左节点的左孩子，右节点的右孩子。
  - 比较二叉树内侧是否对称，传入左节点的右孩子，右节点的左孩子
  - 如果相等，返回True

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        递归 后序遍历
        """
        if not root:
            return True 
            
        
        def compare(left,right):
            if left is not None and right is None :
                return False 
            elif left is None and right is not None :
                return False 
            elif right is None and left is None:
                return True 
            elif right.val != left.val:
                return False 
            # 比较外侧节点是否对称
            outside = compare(left.left,right.right)
            # 比较内侧节点是否对称
            inside = compare(left.right,right.left)
            res = outside and inside 
            return res 
        return compare(root.left,root.right)

    def isSymmetric_deque(self,root):
        '''迭代法，使用队列 
        这道题目我们也可以使用迭代法，但要注意，这里的迭代法可不是前中后序的迭代写法，因为本题的本质是判断两个树是否是相互翻转的
        ，其实已经不是所谓二叉树遍历的前中后序的关系了。

        这里我们可以使用队列来比较两个树（根节点的左右子树）是否相互翻转，（注意这不是层序遍历
        '''
        if not root:
            return None 
        import collections
        queue = collections.deque()
        # 队列的特点：先进先出
        queue.append(root.left) # 将左子树头节点加入队列
        queue.append(root.right)

        while queue:
            leftNode = queue.popleft()
            rightNode = queue.popleft()
            # 左节点为空，右节点为空则继续
            if leftNode is None and rightNode is None:
                continue
            # 1.左节点不为空或者2.右节点不为空或者3.都不为空但数值不同
            if leftNode is None or rightNode is None or leftNode.val!=rightNode.val:
                return False
            
            # 比较外侧
            queue.append(leftNode.left)
            queue.append(rightNode.right)
            # 比较内侧
            queue.append(leftNode.right)
            queue.append(rightNode.left)
        return True 
    
    def isSymmetric_stack(self,root):
        '''迭代法 使用栈'''
        if not root:
            return None 
        
        # 栈的特点是先进后出
        stack = []
        stack.append(root.left)
        stack.append(root.right)

        while stack:
            leftNode = stack.pop()
            rightNode = stack.pop()
            if leftNode is None and rightNode is None:
                continue 
            
            if leftNode is None or rightNode is None or leftNode.val != rightNode.val:
                return False 

            stack.append(leftNode.left)
            stack.append(rightNode.right)

            stack.append(leftNode.right)
            stack.append(rightNode.left)
        return True 



        
            
        

        
        
```
