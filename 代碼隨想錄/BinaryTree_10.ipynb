# 完全二叉树的节点个数

## 问题
给出一个完全二叉树，求出该树的节点个数。

示例 1：

输入：root = [1,2,3,4,5,6]
输出：6
示例 2：

输入：root = []
输出：0


## 思路
### 普通二叉树

#### 递归
与求二叉树的深度写法类似

#### 迭代
使用层序遍历+记录遍历的节点数量就可以


### 完全二叉树
在完全二叉树中，除了最底层节点可能没填满外，其余每层节点数都达到最大值，并且最下面一层的节点都集中在该层最左边的若干位置。若最底层为第 h 层，则该层包含 1~ 2^(h-1)  个节点。
<img src = "https://img-blog.csdnimg.cn/20200920221638903.png">

完全二叉树只有两种情况
- 满二叉树
  直接用[2^树深度-1]来计算，注意这里根节点深度为1

- 最后一层的叶子节点没有满
  分别递归左孩子，右孩子，递归到某一深度一定会有左孩子或右孩子为满二叉树。
  
## 代码



```python
class Solution(object):
    def countNodes(self, root):
        """
        :type root: TreeNode
        :rtype: int
        递归
        """
        def getNodesNum(root):
            if root is None:
                return 0
            # 先求左子树的节点数量，再求右子树的节点数量，最后取总和再加1
            # 加1的原因是算上当前的中间节点
            leftNum = getNodesNum(root.left)
            rightNum = getNodesNum(root.right)
            treeNum = leftNum+rightNum+1 
            return treeNum
        return getNodesNum(root)
    def countNodes_v2(self,root):
        '''迭代法'''
        import collections
        deque = collections.deque()
        if root is not None:
            deque.append(root)
        res = 0 
        while deque:
            size = len(deque)

            for i in range(size):
                node = deque.popleft()
                res+=1 
                if node.left:
                    deque.append(node.left)
                if node.right:
                    deque.append(node.right)
        return res 
        
    def countNodes_v3(self,root):
        '''利用满二叉树的性质'''
        if root is None:
            return 0 

        left = root.left 
        right  = root.right 
        leftDepth,rightDepth = 0,0
        # 计算左子树的深度
        while left:
            left= left.left
            leftDepth+=1

        # 计算右子树的深度
        while right:
            right = right.right
            rightDepth+=1
        # 如果深度相等，就是满二叉树
        if leftDepth==rightDepth:
            return 2<<leftDepth+1# 注意(2<<1) 相当于2^2，所以leftDepth初始为0

        return self.countNodes_v3(root.left)+self.countNodes_v3(root.right)+1
```
