# 二叉树的最小深度

## 问题
给定一个二叉树，找出其最小深度。

最小深度是从根节点到最近叶子节点的最短路径上的节点数量。

说明: 叶子节点是指没有子节点的节点。

示例:

给定二叉树 [3,9,20,null,null,15,7],

<img src = "https://img-blog.csdnimg.cn/2021020315582586.png">

返回它的最小深度2

## 代码

最小深度是从根节点到最近**叶子节点**的最短路径上的节点的节点数量

叶子节点：左右孩子都为空的节点


求二叉树的最小深度和求二叉树的最大深度的差别主要在于处理左右孩子不为空的逻辑。

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def minDepth(self, root):
        """
        :type root: TreeNode
        :rtype: int
        递归法
        """
        def getDepth(root):
            if root is None:
                return 0 
            left_depth = getDepth(root.left)
            right_depth =getDepth(root.right)
            '''犯错误了，
                - 左子树为空，右子树不为空，说明深度是1+右子树的深度
                - 右子树为空，左子树不为空，最小深度是1+左子树的深度
                - 如果左右子树都不为空，则返回右子树深度最小值+1
                ----------------------------------------
                应该添加如下的代码
                '''
            if root.right is None and root.left != None:
                return 1 + left_depth
            if root.left is None and root.right != None:
                return 1+right_depth
            res = 1+min(left_depth,right_depth)
            return res 
            
        return getDepth(root)

    def minDepth_v3(self,root):
        '''递归法
        '''
        if not root:
            return 0 
        if not root.left and not root.right:
            return 1 

        # 先设置一个最大值，然后用min进行比较
        min_depth = 10**9
        if root.left:
            # 获得左子树的最小高度
            min_depth = min(self.minDepth_v3(root.left),min_depth)
        if root.right:
            # 获得右子树的最小高度
            min_depth = min(self.minDepth_v3(root.left),min_depth)

        return min_depth+1


    def minDepth_v2(self,root):
        '''迭代法
        只有当左右孩子都为空，才说明遍历的最低点
        '''
        if not root:
            return None 
        
        import collections
        que = collections.deque()
        que.append(root)
        res = 1 
        
        while que:
            for _ in range(len(que)):
                node = que.popleft()
                # 当左右孩子都为空，说明是最低的一层，退出
                if node.right is None and node.left is None:
                    return res 
                if node.left is not None:
                    que.append(node.left)
                if node.right is not None:
                    que.append(node.right)
            res +=1 
        return res 
```
